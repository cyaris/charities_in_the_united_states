```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('charity_navigator_clean.csv', index_col = 0)
```

```python
df.drop('charity_url', axis = 1, inplace = True)
```

```python
plt.figure(figsize = (15, 8))

sns.heatmap(df.corr(), cmap = 'coolwarm', annot = True)
plt.title('df.corr()');
```

```python
df_25_corr = df[df['comp_leader_income'].isnull() == False]
df_25_corr.drop(['charity_name', 'city', 'cn_advisory', 'state',
                 'org_category', 'comp_leader_title', 
                 'accountability_score', 'financial_score',
                 'overall_score', 'comp_leader_expense_pct',
                 'payments_to_affiliates'],
                 axis = 1, inplace = True)
df_25_corr.head()
```

```python
sns.pairplot(df_25_corr, hue = 'org_type', palette = 'viridis');
```
